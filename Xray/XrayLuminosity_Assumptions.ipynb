```python
import astropy.constants as c 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def v_orbit(M_stripped,P_day) : 
    
    M_neutron = 1.4
    
    a = (M_stripped) ** (1/3)
    
    b = (1 + M_neutron / M_stripped) ** (1/3)
    
    d = (1 / P_day) ** (1/3)
    
    return 200 * a * b * d
```

```python
plt.figure(figsize=(10,10))
plt.style.use('ggplot')

periods = np.arange(.1,10,.1)

plt.scatter(periods,v_orbit(6,periods),label="6")
plt.scatter(periods,v_orbit(8,periods),label="8")
plt.scatter(periods,v_orbit(10,periods),label="10")
plt.scatter(periods,v_orbit(50,periods),label="50")
plt.scatter(periods,v_orbit(100,periods),label="100")
plt.legend(fontsize=18)
plt.ylabel("v orbit (km /s)",fontsize=20)
plt.xlabel("Period (days)",fontsize=20)
plt.ylim(0,1100)
```
