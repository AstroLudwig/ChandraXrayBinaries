```python
df = pd.read_csv("CandidateList.csv")
df
```

```python
def bash_command(cmd,print_output=True):
        output = subprocess.check_output(['/bin/bash','-i', '-c',cmd])
        if print_output:
            print(output)
        return output
```

```python
ni_output = []

for i in range(df.shape[0]) :
    ra = df.ra.iloc[i]
    dec = df.dec.iloc[i]
    output = bash_command('heainit; nh equinox=2000 ra=%s dec=%s' % (ra, dec),print_output=False)
    ni_output.append(output.decode("utf-8").split('\n'))
```

```python
np.savetxt('NH.txt',ni_output, fmt='%s')
```
